```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

# Load the dataset
data = pd.read_csv('https://cocl.us/concrete_data')

# Assuming the target column is named 'concrete_strength'. 
# Please replace it with the actual column name for concrete strength in your dataset
X_data = data.drop('Strength', axis=1)
Y_data = data['Strength']

# Normalize the data
scaler = StandardScaler()
X_data = scaler.fit_transform(X_data)

mean_squared_errors = []

for i in range(50):
    # Splitting the data
    X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size=0.3)

    # Defining the model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(1))

    # Compiling the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Training the model
    model.fit(X_train, y_train, epochs=50, verbose=0)

    # Predicting using the model
    y_pred = model.predict(X_test)

    # Computing the mean squared error
    mse = mean_squared_error(y_test, y_pred)

    # Appending to the list
    mean_squared_errors.append(mse)

# Converting list to numpy array
mean_squared_errors = np.array(mean_squared_errors)

# Reporting the mean and standard deviation of the mean squared errors
print("Mean of Mean Squared Errors: ", np.mean(mean_squared_errors))
print("Standard Deviation of Mean Squared Errors: ", np.std(mean_squared_errors))
```
